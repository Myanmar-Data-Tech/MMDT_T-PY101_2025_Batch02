```python
#Q4) The following program should check if a number is a prime number. Debug the code.

def is_prime(n):
    for num in range(2, n):
        if n % num == 0:
            print(f"No Prime ={n}, num={num}")
            return False
    print(f"Prime : n={n}")
    return True

numbers = np.arange(3, 21)
results = {}
for num in numbers:
    results[int(num)] =is_prime(int(num))

print(f"Your results is {results}")
expected_result = {3: True, 4: False, 5: True, 6: False, 7: True, 8: False, 9: False, 10: False, 11: True, 12: False, 13: True, 14: False, 15: False, 16: False, 17: True, 18: False, 19: True, 20: False}
print(f"Expected result is: {expected_result}")
```
