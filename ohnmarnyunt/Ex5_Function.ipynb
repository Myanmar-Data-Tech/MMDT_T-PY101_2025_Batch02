Ex 3: Write a function called number_of_factors that takes an integer and returns how many
factors the number has.

```python
def number_of_factors(dividend):
  factors = []
  for divisor in range(1,dividend+1):
    if (dividend%divisor==0):
      factors.append(divisor)
  return factors

num = int(input("Enter an integer: "))
factors = number_of_factors(num)
print(f"The factors of {num} are: {factors}. \nThe number of factors for the given integer is {len(factors)}")
```

Ex7:  Write a function called change_case that given a string, returns a string with each upper  case letter replaced by a lower case letter and vice-versa

```python
def change_case(my_str):
  vice_versa=""
  for char in my_str:
    if str.isupper(char):
      vice_versa += char.lower()
    elif str.islower(char):
      vice_versa += char.upper()
    else:
      vice_versa += char
  return vice_versa

print(change_case("Hello World!"))
print(change_case("HELLO python!"))
```
