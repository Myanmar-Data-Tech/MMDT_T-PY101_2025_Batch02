Ex 3: Write a function called number_of_factors that takes an integer and returns how many
factors the number has.

```python
def number_of_factors(n):
  count = 0
  for i in range(1,n+1):
    if n % i == 0:
      count = count + 1
  return count

print(number_of_factors(7))
```

```python
def first_diff(m,n):
  m1 = m.upper()
  n1 = n.upper()
  r = min(len(m1),len(n1))
  for i in range(r):
    if m1[i] != n1[i]:
      return i
  if m1[i] == n1[i]:
    return -1


print(first_diff('Impossible','impossibel'))
```

Ex6: Write a function that solves the system of equations ax + by = e and cx +dy = f. Your function should takes 6 parameters, (a, b, c, d, e,f) and return the values of x and y.

```python
#using Cramer's Rule
def solves(a,b,c,d,e,f):
  det = (a*d) - (b*c)
  if det == 0:
    return "No Solution"
  else:
    x = ((e*d)-(b*f))/det
    y = ((a*f)-(b*c))/det
    return x,y
print(solves(3,3,2,3,12,11))
```

Ex7:  Write a function called change_case that given a string, returns a string with each upper  case letter replaced by a lower case letter and vice-versa

```python
def change_case(s):
  new_s = " "
  for char in s:
    if char.isupper():
      new_s = new_s + char.lower()
    if char.islower():
      new_s = new_s + char.upper()
  return new_s

print(change_case("Hello"))
```

Ex8:  Write a function called primes that is given a number n and returns a list of the first n  primes. Let the default value of n be 100.

```python
def primes(n=100):
  prime_numbers = []
  number = 2
  while len(prime_numbers) < n:
    for i in range(2,number):
      if number % i == 0:
        break
    else:
      prime_numbers.append(number)
    number = number + 1
  return prime_numbers
print(primes(7))
```
